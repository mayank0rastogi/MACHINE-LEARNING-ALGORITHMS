```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```
data=pd.read_csv('advertising.csv')
data.head()
```

```
X=data[['Age','EstimatedSalary']]
Y=data['Purchased']
```

```
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

###Apply Visualization on the  data

```
import seaborn as sns
sns.pairplot(data,hue='Purchased')
```

```
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

###KNN Classifier Model Fit

```
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

```
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
metrics.accuracy_score(y_test,y_pred)
```

**Applying cross_val_score**

```
from sklearn.model_selection import cross_val_score
```

```
knn=KNeighborsClassifier(n_neighbors=4)
cross_val_score(knn,X,Y,cv=10,scoring='accuracy').mean()
```

###APPLY ON LOGISTIC REGRESSION

```
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
cross_val_score(logreg, X, Y, cv=10, scoring = 'accuracy').mean()
```

####APPLY ON DECISION TREE CLASSIFIER

```
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeClassifier(criterion='entropy',splitter='random')
print (cross_val_score(tree, X,Y, cv=10, scoring = 'accuracy').mean())
```

```
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeClassifier()
print (cross_val_score(tree, X,Y, cv=10, scoring = 'accuracy').mean())
```
